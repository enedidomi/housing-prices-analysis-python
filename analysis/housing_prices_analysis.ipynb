```python
# 2. Data visualization

# 2.1 Distribution of house prices

plt.figure(figsize=(8,5))
plt.hist(mydata['sale_price'], bins=30)
plt.title("Distribution of House Prices")
plt.xlabel("Sale Price ($)")
plt.ylabel("Number of Houses")
plt.show()
```

```python
# 2.2 Boxplot of house prices

plt.figure(figsize=(8,5))
plt.boxplot(mydata['sale_price'], vert=False, showfliers=False)
plt.title("Boxplot of House Prices")
plt.ylabel("Sale Price ($)")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 2.3 Living area vs. sale price scatter plot

plt.figure(figsize=(8,5))
plt.scatter(mydata['living_area'], mydata['sale_price'], alpha=0.5)
plt.title("Living Area vs. Sale Price")
plt.xlabel("Living Area (sq ft)")
plt.ylabel("Sale Price ($)")
plt.show()
```

```python
# 2.4 Average sale price by number of bedrooms in a bar chart

avg_price_x_bedrooms = mydata[mydata['bedrooms'] > 0].groupby('bedrooms')['sale_price'].mean()

plt.figure(figsize=(8,5))
plt.bar(avg_price_x_bedrooms.index, avg_price_x_bedrooms.values)
plt.title("Average Sale Price by Number of Bedrooms")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Average Sale Price ($)")
plt.xticks(avg_price_x_bedrooms.index)
plt.show()
```

```python
# 2.5 Average sale price by overall quality in a bar chart

avg_price_x_quality = mydata.groupby('overall_quality')['sale_price'].mean()

plt.figure(figsize=(8,5))
plt.bar(avg_price_x_quality.index, avg_price_x_quality.values)
plt.title("Average Sale Price by Overall Quality")
plt.xlabel("Overall Quality")
plt.ylabel("Average Sale Price ($)")
plt.xticks(avg_price_x_quality.index)
plt.show()
```

```python
# 2.6 Average sale price by year built in a line chart

avg_price_x_yearbuilt = mydata.groupby('year_built')['sale_price'].mean()

plt.figure(figsize=(12,5))
plt.plot(avg_price_x_yearbuilt.index, avg_price_x_yearbuilt.values)
plt.title("Average Sale Price by Year Built")
plt.xlabel("Year Built")
plt.ylabel("Average Sale Price ($)")
plt.show()
```

```python
# 3. Correlation analysis

# 3.1 Create the correlation matrix

correlation_matrix = mydata.corr(numeric_only=True)

correlation_matrix
```

```python
# 3.2 Correlation with sale price

sale_price_corr = correlation_matrix['sale_price'].sort_values(ascending=False)

sale_price_corr
```

```python
# 3.3 Correlation visualisations

price_correlations = sale_price_corr.drop('sale_price')

plt.figure(figsize=(8,8))
plt.barh(price_correlations.index, price_correlations.values)
plt.title("Correlation with Sale Price")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Variables")
plt.grid(axis='x', linestyle='--', alpha=0.3)
plt.show()
```

```python
# 4. Simple regression analysis

x = mydata['living_area']
y = mydata['sale_price']

# Calculate regression line
slope, intercept = np.polyfit(x, y, 1)

plt.figure(figsize=(8,5))
plt.scatter(x, y, alpha=0.5)
plt.plot(x, slope * x + intercept, color='red', linewidth=2)
plt.title("Living Area vs. Sale Price with Regression Line")
plt.xlabel("Living Area (sq ft)")
plt.ylabel("Sale Price ($)")
plt.show()
```
