```python
# Remove columns with many missing values

mydata = mydata.drop(columns=[
    'pool_quality',
    'misc_feature'
])
```

```python
# Fill missing categorical values step 1

mydata['alley_type'] = mydata['alley_type'].fillna('Not Available')
mydata['masonry_veneer_type'] = mydata['masonry_veneer_type'].fillna('Not Available')
mydata['fireplace_quality'] = mydata['fireplace_quality'].fillna('Not Available')
mydata['fence'] = mydata['fence'].fillna('Not Available')
```

```python
# Fill missing categorical values step 2

mydata['basement_quality'] = mydata['basement_quality'].fillna('Not Available')
mydata['basement_condition'] = mydata['basement_condition'].fillna('Not Available')
mydata['basement_exposure'] = mydata['basement_exposure'].fillna('Not Available')
mydata['basement_finish_type_1'] = mydata['basement_finish_type_1'].fillna('Not Available')
mydata['basement_finish_type_2'] = mydata['basement_finish_type_2'].fillna('Not Available')
```

```python
# Fill missing categorical values step 3

mydata['garage_type'] = mydata['garage_type'].fillna('Not Available')
mydata['garage_finish'] = mydata['garage_finish'].fillna('Not Available')
mydata['garage_quality'] = mydata['garage_quality'].fillna('Not Available')
mydata['garage_condition'] = mydata['garage_condition'].fillna('Not Available')
```

```python
# Replace missing numerical values with 0 step 1

mydata['garage_year_built'] = mydata['garage_year_built'].fillna(0)
```

```python
# Replace missing numerical values with 0 step 2

mydata['masonry_veneer_area'] = mydata['masonry_veneer_area'].fillna(0)
```

```python
# Fill missing values with the mode

mydata['electrical_system'] = mydata['electrical_system'].fillna(
    mydata['electrical_system'].mode()[0]
)
```

```python
# Fill missing values with the mean

mydata['lot_frontage'] = mydata['lot_frontage'].fillna(
    mydata['lot_frontage'].median()
)
```

```python
# Count duplicate entries

mydata.duplicated().sum()
```

```python
# Filter numeric variables and check for negative values

numeric_columns = mydata.select_dtypes(include=['int64', 'float64']).columns

(mydata[numeric_columns] < 0).sum()
```
